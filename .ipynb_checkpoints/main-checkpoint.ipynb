```python
import numpy as np
import matplotlib.lines as mlines
from matplotlib import pyplot as plt
```

```python
# fxn for plotting
def get_ax(figsize=(6,4)):
    fig, ax = plt.subplots(figsize=figsize)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    return ax
```

```python
# Ec = e^2/2C
# El = (hbar/2e)^2/L
# El << Ej
# 1 <~ Ej/Ec <~ 10

def H(n, phi, phi_ext, Ec, El, Ej):
    return (4*Ec*n**2) + (0.5*El*phi**2) - Ej*np.cos(phi - phi_ext)
```

```python
C = 1
L = 1000
Ec = 1/(2*C)
El = 1/(4*L)
Ej = 1
phi_ext = 0
n = 1
print(f"Ec: {Ec}, El: {El}, Ej: {Ej}")
print("Ej/Ec:", Ej/Ec)
```

```python
phi_list = np.arange(0,2*np.pi,0.1)
H_list = H(n, phi_list, phi_ext, Ec, El, Ej)
```

```python
plt.plot(phi_list, H_list)
```
